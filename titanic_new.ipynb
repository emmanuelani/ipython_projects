```python
import pandas as pd
import numpy as np
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df_1 = df[['Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
df = df_1

df.head()
```

```python
target = df['Survived']
inputs = df.drop('Survived', axis=1)

# Converting the categorical variables to dummy variables using pd.get_dummies
dummies = pd.get_dummies(inputs['Sex'])
dummies

# Concatenating the dummy variable with the input dataframe
inputs = pd.concat([inputs, dummies], axis=1)
# Dropping the Sex column
inputs = inputs.drop('Sex', axis=1)
inputs.isnull().sum()

# replacing Age null values with the mean age
mean_age = inputs['Age'].mean()
# inputs.replace(np.NaN, mean_age)
inputs['Age'].fillna(mean_age, inplace=True)

inputs.columns[inputs.isna().any()]
# inputs.isna().any()

inputs['Age'][:10]
mean_age = inputs['Age'].mean()
mean_age

# df.query('Age==29')
```

```python
# splitiing the dataset 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score

X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)

reg = LogisticRegression()

reg.fit(X_train, y_train)
reg.score(X_train, y_train)

y_pred = reg.predict(X_test)
confusion_matrix(y_test, y_pred)

f1_score(y_test, y_pred)

y_pred = pd.Series(y_pred)
# pd.concat(y_test, y_pred)

# y_pred = y_pred.to_frame()
# y_test = y_test.to_frame()
print('Length of X_train:', len(X_train))
print('Length of X_test:', len(X_test))

```

```python
# Using naive bayes classifier
from sklearn.naive_bayes import GaussianNB # for normally distributed data
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

model = GaussianNB()

model.fit(X_train, y_train)
model.score(X_test, y_test)

precision_score(y_test, y_pred)
recall_score(y_test, y_pred)
```
